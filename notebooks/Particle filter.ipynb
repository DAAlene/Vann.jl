# Particle filter

## Test likelihood function

```julia
using Distributions
using PyPlot
```

```julia
q_sim = collect(0:0.01:50);
q_obs = 10.;
```

```julia
y = [];

for q in q_sim
    sigma = 0.1 * q_obs;
    sigma = max(sigma, 0.1);
    res = 1 / sqrt(2*pi*sigma^2) * exp( -(q-q_obs)^2 / (2*sigma^2) );
    push!(y, res);
end
```

```julia
plot(q_sim, y);
```

```julia
y = [];

for q in q_sim
    y = push!(y, pdf(Normal(q_obs, max(0.5*q_obs, 0.5)), q) );
end

plot(q_sim, y);
```
